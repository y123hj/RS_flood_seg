```python
import os
import random
```

```python
train_ratio = 0.8
test_ratio = 1 - train_ratio

rootdata = '/content/drive/MyDrive/RS/HAND/S1_split/'

imagelist = os.listdir(rootdata)
random.shuffle(imagelist)
```

```python
train_list = []
test_list = []
for i in range(0,int(len(imagelist)*train_ratio)):
    train_list.append(imagelist[i])
for i in range(int(len(imagelist)*train_ratio),len(imagelist)):
    test_list.append(imagelist[i])
```

```python
print(len(train_list))
print(len(test_list))
print(train_list)
```

```python
train_list.sort()
test_list.sort()
with open('/content/drive/MyDrive/RS/HAND/train.txt', 'w') as f:
    for item in train_list:
        f.write("%s\n" % item)

with open('/content/drive/MyDrive/RS/HAND/test.txt', 'w') as f:
    for item in test_list:
        f.write("%s\n" % item)
```
